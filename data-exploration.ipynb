# Data Exploration

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
starb = pd.read_csv("starbucks.csv")
```

```python
starb.head()
```

```python
# more than 100, small, tall, grande, venti
filtered = starb[(starb['size'].isin(["short", "tall", "grande", "venti"])) & (starb['caffeine_mg']>100)]
```

```python
# filtered.to_csv("filtered_starbucks.csv")
```

```python
sns.scatterplot(x="caffeine_mg", y="sugar_g", alpha=0.6, data=filtered[filtered['size']=="venti"])
plt.axvline(x=400, color='r', label='max-suggestion')
plt.axhline(y=50, color='r', label='max-suggestion')
plt.title("Sugar and Caffeine Content in Venti Starbucks Drinks")
plt.show()
```

```python
sns.histplot(x="caffeine_mg", hue="size", alpha=0.6, multiple="stack", palette="Pastel2",
             data=filtered)
plt.axvline(x=400, color='r', label='max-suggestion')
plt.title("Distribution of Caffeine Content in Starbucks Drinks by Size")
plt.show()
```

```python
sns.histplot(x="sugar_g", hue="size", alpha=0.6, multiple="stack", palette="Pastel2",
             data=filtered)
plt.axvline(x=50, color='r', label='max-suggestion')
plt.title("Distribution of Sugar Content in Starbucks Drinks by Size")
plt.show()
```
